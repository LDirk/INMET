```python
caminho_2024 = 'DADOS_INMET/2024'
caminho_2023 = 'DADOS_INMET/2023'
caminho_2022 = 'DADOS_INMET/2022'
caminho_2021 = 'DADOS_INMET/2021'
caminho_2020 = 'DADOS_INMET/2020'
caminho_2019 = 'DADOS_INMET/2019'

#caminho_2018 = 'DADOS_INMET/2018'

```

```python
arquivos_pasta2024 = [f for f in os.listdir(caminho_2024) if os.path.isfile(os.path.join(caminho_2024, f))]
arquivos_pasta2023 = [f for f in os.listdir(caminho_2023) if os.path.isfile(os.path.join(caminho_2023, f))]
arquivos_pasta2022 = [f for f in os.listdir(caminho_2022) if os.path.isfile(os.path.join(caminho_2022, f))]
arquivos_pasta2021 = [f for f in os.listdir(caminho_2021) if os.path.isfile(os.path.join(caminho_2021, f))]
arquivos_pasta2020 = [f for f in os.listdir(caminho_2020) if os.path.isfile(os.path.join(caminho_2020, f))]
arquivos_pasta2019 = [f for f in os.listdir(caminho_2019) if os.path.isfile(os.path.join(caminho_2019, f))]

#arquivos_pasta2018 = [f for f in os.listdir(caminho_2018) if os.path.isfile(os.path.join(caminho_2018, f))]



```

```python
estacoes = {
    'INMET_SE_RJ_A602_RIO DE JANEIRO-MARAMBAIA': 'Marambaia',
    'INMET_SE_RJ_A602_MARAMBAIA': 'Marambaia',
    'INMET_SE_RJ_A621_RIO DE JANEIRO - VILA MILITAR': 'Vila Militar',
    'INMET_SE_RJ_A621_VILA MILITAR': 'Vila Militar',
    'INMET_SE_RJ_A627_NITEROI': 'Niteroi',
    'INMET_SE_RJ_A636_RIO DE JANEIRO - JACAREPAGUA': 'Jacarepagua',
    'INMET_SE_RJ_A652_RIO DE JANEIRO - FORTE DE COPACABANA': 'Forte de Copacabana',
    'INMET_SE_RJ_A652_FORTE DE COPACABANA': 'Forte de Copacabana',
}
```

```python

# Função para processar arquivos e combinar dados
def processar_arquivos(caminho, arquivos):
    df_total = pd.DataFrame()
    for arquivo in arquivos:
        # Criar a chave do mapeamento com base no nome do arquivo (até o último "_")
        nome_base = '_'.join(arquivo.split('_')[:5])
        estacao = estacoes.get(nome_base)
        if not estacao:
            raise ValueError(f"Estação não encontrada para o arquivo: {arquivo}")
        
        # Montar o caminho completo do arquivo
        caminho_arquivo = os.path.join(caminho, arquivo)
        
        # Ler o arquivo CSV
        try:
            df_temp = pd.read_csv(
                caminho_arquivo,
                sep=';',
                encoding='ISO-8859-1',
                skiprows=8,
                on_bad_lines='skip'
            )
            df_temp['ESTACAO'] = estacao
            df_total = pd.concat([df_total, df_temp], ignore_index=True)
        except FileNotFoundError:
            print(f"Arquivo não encontrado: {arquivo}")
        except Exception as e:
            print(f"Erro ao processar {arquivo}: {e}")
    return df_total

```

```python
# Processar dados 
df_total_2024 = processar_arquivos(caminho_2024, arquivos_pasta2024)
df_total_2023 = processar_arquivos(caminho_2023, arquivos_pasta2023)
df_total_2022 = processar_arquivos(caminho_2022, arquivos_pasta2022)
df_total_2021 = processar_arquivos(caminho_2021, arquivos_pasta2021)
df_total_2020 = processar_arquivos(caminho_2020, arquivos_pasta2020)
df_total_2019 = processar_arquivos(caminho_2019, arquivos_pasta2019)

df_total_2018 = processar_arquivos(caminho_2018, arquivos_pasta2018)


df_2019_ate_2024 = pd.concat([df_total_2024,df_total_2023, df_total_2022,df_total_2021,df_total_2020,df_total_2019   ], ignore_index=True)


```
